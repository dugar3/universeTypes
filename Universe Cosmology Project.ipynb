```python
import numpy as np
from astropy import units as u
from astropy import constants as const
from astropy.cosmology import LambdaCDM
import matplotlib.pyplot as plt
```

```python
#Om0 = Omega_m (matter), Ode0 = Omega_lambda (dark energy), Ogamma0 = Omega_r (radiation)
z = np.linspace(0, 6, 1000);

cosmo1 = LambdaCDM(H0=70, Om0=1, Ode0=0);
cosmo2 = LambdaCDM(H0=70, Om0=0, Ode0=0, Tcmb0 = 32.336572219157243);
cosmo3 = LambdaCDM(H0=70, Om0=0, Ode0=1);
cosmo4 = LambdaCDM(H0=70, Om0=0.3, Ode0=0);
cosmo5 = LambdaCDM(H0=70, Om0=0.3, Ode0=0.7);

prop_dist1 = cosmo1.comoving_distance(z)*70/(3*10**5);
prop_dist2 = cosmo2.comoving_distance(z)*70/(3*10**5);
prop_dist3 = cosmo3.comoving_distance(z)*70/(3*10**5);
prop_dist4 = cosmo4.comoving_distance(z)*70/(3*10**5);
prop_dist5 = cosmo5.comoving_distance(z)*70/(3*10**5);

plt.plot(z,prop_dist1,label='Flat, Matter-only')
plt.plot(z,prop_dist2,label='Flat, Radiation-only')
plt.plot(z,prop_dist3,label='Flat, $\Lambda$-only')
plt.plot(z,prop_dist4,label='Open')
plt.plot(z,prop_dist5,label='Consensus model, flat')

plt.xlabel('Red Shift (z)')
plt.ylabel('Proper Distance (c/$H_0$)')
plt.legend()
plt.show
```

```python
lum_dist1 = cosmo1.luminosity_distance(z)*70/(3*10**5);
lum_dist2 = cosmo2.luminosity_distance(z)*70/(3*10**5);
lum_dist3 = cosmo3.luminosity_distance(z)*70/(3*10**5);
lum_dist4 = cosmo4.luminosity_distance(z)*70/(3*10**5);
lum_dist5 = cosmo5.luminosity_distance(z)*70/(3*10**5);

plt.plot(z,lum_dist1,label='Flat, Matter-only')
plt.plot(z,lum_dist2,label='Flat, Radiation-only')
plt.plot(z,lum_dist3,label='Flat, $\Lambda$-only')
plt.plot(z,lum_dist4,label='Open')
plt.plot(z,lum_dist5,label='Consensus model, flat')

plt.xlabel('Red Shift (z)')
plt.ylabel('Luminosity Distance (c/$H_0$)')
plt.legend()
plt.show
```
